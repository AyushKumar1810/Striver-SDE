Instructions:
Generate Data:
Use numpy to generate a set of 100 random values for x in the range 0 to 10.
Use numpy to generate a set of 100 random values for y that are approximately linearly related to x with some noise.
Create a DataFrame:
Using pandas, create a DataFrame that contains the x and y values.
Plot a Scatterplot:
Use matplotlib to plot a scatterplot of the x and y values.
Label the x-axis as "X values" and the y-axis as "Y values".
Title the plot as "Scatterplot of X vs Y".

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.uniform(0, 10, 100)
y = 2.5 * x + np.random.normal(0, 1, 100)

df = pd.DataFrame({'X': x, 'Y': y})

plt.scatter(df['X'], df['Y'])
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatterplot of X vs Y')
plt.show()
```
